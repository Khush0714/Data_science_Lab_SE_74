```python
def is_odd_or_even(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
num1 = 4
print(f"The number {num1} is {is_odd_or_even(num1)}.")

num2 = 7
print(f"The number {num2} is {is_odd_or_even(num2)}.")

num3 = 0
print(f"The number {num3} is {is_odd_or_even(num3)}.")

num4 = -3
print(f"The number {num4} is {is_odd_or_even(num4)}.")
```

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reversed version
    return num_str == num_str[::-1]

# Example usage:
num1 = 121
print(f"Is {num1} a palindrome? {is_palindrome(num1)}")

num2 = 123
print(f"Is {num2} a palindrome? {is_palindrome(num2)}")

num3 = 12321
print(f"Is {num3} a palindrome? {is_palindrome(num3)}")

num4 = -121 # Negative numbers are typically not considered palindromes in their numerical form
print(f"Is {num4} a palindrome? {is_palindrome(num4)}") # Will return False based on string comparison
```

```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_display = 7
display_multiplication_table(num_to_display)

print("\n") # Add a newline for better readability

num_to_display = 12
display_multiplication_table(num_to_display)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = 7
print(f"The factorial of {number} is {factorial(number)}")
```

## Define the Problem and Rules

### Decision-Making Problem for Movie Recommendation Expert System

The goal is to build an expert system that recommends a movie genre based on two primary input parameters from the user:

1.  **User Mood**: The current emotional state of the user (e.g., happy, sad, excited, relaxed, thoughtful, scared).
2.  **Time of Day**: The current time period (e.g., morning, afternoon, evening, night).

The system will process these inputs using a set of predefined rules to output a suitable movie genre recommendation.

### Rules for Movie Recommendations:

Here are the rules that define the expert system's logic:

*   **Rule 1**: If **Mood** is 'happy' AND **Time of Day** is 'evening', recommend 'Comedy'.
*   **Rule 2**: If **Mood** is 'sad' AND **Time of Day** is 'night', recommend 'Drama' or 'Comfort Movie'.
*   **Rule 3**: If **Mood** is 'adventurous' AND **Time of Day** is 'afternoon', recommend 'Action/Adventure'.
*   **Rule 4**: If **Mood** is 'relaxed' AND **Time of Day** is 'morning', recommend 'Documentary' or 'Light Drama'.
*   **Rule 5**: If **Mood** is 'excited' AND **Time of Day** is 'evening', recommend 'Thriller' or 'Action'.
*   **Rule 6**: If **Mood** is 'thoughtful' AND **Time of Day** is 'any', recommend 'Mystery' or 'Sci-Fi'.
*   **Rule 7**: If **Mood** is 'scared' AND **Time of Day** is 'night', recommend 'Horror' (for brave souls) OR 'Animated Family' (to counter fear).

## Design and Implement Expert System

### Subtask:
Write Python code to implement the expert system, including functions to gather user input (e.g., mood, time) and an inference engine that applies the predefined rules to provide a decision or recommendation.


**Reasoning**:
I will implement the expert system by defining two functions: one to gather and validate user input for mood and time of day, and another to apply the predefined recommendation rules. Finally, I will call these functions to demonstrate the system's decision-making capabilities.



```python
def get_user_input():
    """
    Prompts the user for their mood and time of day, with input validation.
    Returns: tuple (str, str) - (mood, time_of_day)
    """
    valid_moods = ['happy', 'sad', 'adventurous', 'relaxed', 'excited', 'thoughtful', 'scared']
    valid_times = ['morning', 'afternoon', 'evening', 'night']

    mood = ""
    time_of_day = ""

    while mood not in valid_moods:
        mood = input(f"Enter your current mood ({', '.join(valid_moods)}): ").lower()
        if mood not in valid_moods:
            print("Invalid mood. Please choose from the list.")

    while time_of_day not in valid_times:
        time_of_day = input(f"Enter the time of day ({', '.join(valid_times)}): ").lower()
        if time_of_day not in valid_times:
            print("Invalid time of day. Please choose from the list.")

    return mood, time_of_day

def recommend_movie(mood, time_of_day):
    """
    Recommends a movie genre based on user's mood and time of day using predefined rules.
    Args:
        mood (str): The user's current emotional state.
        time_of_day (str): The current time period.
    Returns: str - The recommended movie genre.
    """
    recommendation = "I couldn't find a specific recommendation for your mood and time combination."

    # Rule 1: If Mood is 'happy' AND Time of Day is 'evening', recommend 'Comedy'.
    if mood == 'happy' and time_of_day == 'evening':
        recommendation = 'Comedy'
    # Rule 2: If Mood is 'sad' AND Time of Day is 'night', recommend 'Drama' or 'Comfort Movie'.
    elif mood == 'sad' and time_of_day == 'night':
        recommendation = 'Drama or Comfort Movie'
    # Rule 3: If Mood is 'adventurous' AND Time of Day is 'afternoon', recommend 'Action/Adventure'.
    elif mood == 'adventurous' and time_of_day == 'afternoon':
        recommendation = 'Action/Adventure'
    # Rule 4: If Mood is 'relaxed' AND Time of Day is 'morning', recommend 'Documentary' or 'Light Drama'.
    elif mood == 'relaxed' and time_of_day == 'morning':
        recommendation = 'Documentary or Light Drama'
    # Rule 5: If Mood is 'excited' AND Time of Day is 'evening', recommend 'Thriller' or 'Action'.
    elif mood == 'excited' and time_of_day == 'evening':
        recommendation = 'Thriller or Action'
    # Rule 6: If Mood is 'thoughtful' AND Time of Day is 'any', recommend 'Mystery' or 'Sci-Fi'.
    elif mood == 'thoughtful': # Time of Day is 'any' for thoughtful
        recommendation = 'Mystery or Sci-Fi'
    # Rule 7: If Mood is 'scared' AND Time of Day is 'night', recommend 'Horror' (for brave souls) OR 'Animated Family' (to counter fear).
    elif mood == 'scared' and time_of_day == 'night':
        recommendation = 'Horror (for brave souls) OR Animated Family (to counter fear)'

    return recommendation

# --- Example Usage ---
print("Welcome to the Movie Recommendation Expert System!")

# Get user input
user_mood, user_time = get_user_input()

# Get recommendation
movie_recommendation = recommend_movie(user_mood, user_time)

# Print the recommendation
print(f"Based on your mood ('{user_mood}') and the time of day ('{user_time}'), we recommend: {movie_recommendation}")

print("\n--- Testing with another scenario ---")
# Test scenario 2: Happy and morning (no specific rule match)
user_mood_test = 'happy'
user_time_test = 'morning'
movie_recommendation_test = recommend_movie(user_mood_test, user_time_test)
print(f"For mood '{user_mood_test}' and time '{user_time_test}': {movie_recommendation_test}")

print("\n--- Testing with a specific rule match ---")
# Test scenario 3: Thoughtful (should hit Rule 6)
user_mood_test_2 = 'thoughtful'
user_time_test_2 = 'afternoon'
movie_recommendation_test_2 = recommend_movie(user_mood_test_2, user_time_test_2)
print(f"For mood '{user_mood_test_2}' and time '{user_time_test_2}': {movie_recommendation_test_2}")
```
